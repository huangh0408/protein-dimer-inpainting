```python
import ipdb
import os
from glob import glob
import pandas as pd
import numpy as np
import cv2
from model import *
from util import *
import sys
```

```python
dataset_path="./datasets"
data_type="contact"
data_pickle="./data"
pretrained_model_path="./models"
net_size=256
testset_path  = os.path.join(data_pickle,'protein_testset_'+'_'+data_type+'_size'+str(net_size)+'.pickle')
testset_dir = os.path.join( dataset_path, data_type)
testset_chain_dir = os.path.join( dataset_path, "length" )
testset_dir_temp=os.listdir(testset_dir)
testset_dir_temp.sort()
testset_chain_dir_temp=os.listdir(testset_chain_dir)
testset_chain_dir_temp.sort()
testset = pd.DataFrame({'image_path': map(lambda x: os.path.join( testset_dir, x ), testset_dir_temp),'chain_path':map(lambda x: os.path.join( testset_chain_dir, x ), testset_chain_dir_temp)})

testset.to_pickle( testset_path )
```

```python
batch_size=3
overlap_size = 0
hiding_size = net_size/2
```

```python
is_train = tf.placeholder( tf.bool )
images_tf = tf.placeholder( tf.float32, [batch_size, net_size, net_size, 3], name="images")
images_global = tf.placeholder( tf.float32, [batch_size, net_size, net_size, 3], name="images_global")
ll=tf.placeholder( tf.int32, [batch_size])
model = Model()
bn1, bn2, bn3, bn4, bn5, bn6, debn4, debn3, debn2, debn1, reconstruction_ori, reconstruction=model.build_reconstruction(images_tf,images_global,ll,net_size, is_train)
```

```python
os.environ["CUDA_VISIBLE_DEVICES"]="-1"
sess = tf.Session()
model_file=tf.train.latest_checkpoint(pretrained_model_path)
restorer = tf.train.Saver()
restorer.restore( sess, model_file )
```

```python
test_image_paths = testset[:batch_size]['image_path'].values
test_chain_paths = testset[:batch_size]['chain_path'].values
test_chain_ori=map(lambda x: load_chain( x,net_size ), test_chain_paths)
test_images_ori = map(lambda x: load_image(x,net_size), test_image_paths)
test_images_crop = map(lambda x,y: crop_interaction(x, y,net_size), test_images_ori,test_chain_ori)
test_images,test_global,test_crops, xs,ys = zip(*test_images_crop)
```

```python
bn1_val,bn2_val,bn3_val,bn4_val,bn5_val,bn6_val,debn4_val, debn3_val, debn2_val, debn1_val, reconstruction_vals, recon_ori_vals = sess.run(
	[bn1,bn2,bn3,bn4,bn5,bn6,debn4, debn3, debn2, debn1, reconstruction_ori,reconstruction],
        feed_dict={
        	images_tf: test_images,
#                images_hiding: test_crops,
                is_train: False
                })
```

```python
def plot_input_output_of_this_protein(Input,Output_con,True_con,name):
    from matplotlib.pyplot import figure
    figure(num=None, figsize=(16, 16), dpi=80, facecolor='w', frameon=True, edgecolor='k')
    plt.subplot(1, 3, 1)
    plt.grid(None)
    plt.imshow(Input, cmap='gray', interpolation='nearest')
    plt.title("Input")
    plt.subplot(1, 3, 2)
    plt.grid(None)
    plt.imshow(Output_con, cmap='gray', interpolation='nearest')
    plt.title("Output_contact")
    #plt.subplot(1, 4, 3)
    #plt.grid(None)
    #plt.imshow(Output_dis, cmap='RdYlBu', interpolation='nearest')
    #plt.title("Output_distance")
    plt.subplot(1, 3, 3)
    plt.grid(None)
    plt.imshow(True_con, cmap='RdYlBu', interpolation='nearest')
    plt.title("True_contact")
   
    plt.suptitle(name+'_Prediction')
    plt.show()
```

```python
iii=0
for rec_val,img,crop,x,y in zip(reconstruction_vals,test_images,test_global, xs, ys):
    rec_hid = (255. * rec_val).astype(int)
    rec_con = (255. * img).astype(int)
    #crops=(255. * crop).astype(int)
    rec_1=rec_con.copy()
    xx=net_size-x
    rec_2=misc.imresize(rec_hid[:,:,0],[y,xx],interp='nearest')
    rec_3=misc.imresize(rec_hid[:,:,1],[y,xx],interp='nearest')
    rec_4=misc.imresize(rec_hid[:,:,2],[y,xx],interp='nearest')
    rec_hid_temp=rec_1[0:y,x:net_size]
    rec_hid_temp[:,:,0]=rec_2
    rec_hid_temp[:,:,1]=rec_3
    rec_hid_temp[:,:,2]=rec_4
    rec_con[0:y, x:net_size] = rec_hid_temp

    name=test_image_paths[iii]
    name_temp=name.split('/')[-1]
    pdb_name=name_temp.split('.')[0]
    Input=img.copy()
    Input[0:y,x:net_size,0]=0
    Input[0:y,x:net_size,1]=0
    Input[0:y,x:net_size,2]=0
    Output_con=rec_con.copy()
    Output_con=Output_con/255.
    #Output_dis=rec_con.copy()*255.
    True_con=crop.copy()*255.
    #True_con=255.*True_con
    #plt.imshow(rec_con, cmap='RdYlBu', interpolation='nearest')
    plot_input_output_of_this_protein(Input,Output_con,True_con,pdb_name)
    iii += 1
```
